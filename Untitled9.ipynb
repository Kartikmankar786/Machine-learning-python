```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('your_filename.csv')
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Binning the participant ages
bins = [20, 30, 40, 50, 60]  # Define your bins here
labels = ['20-29', '30-39', '40-49', '50-59']  # Labels for the bins

# Assigning bins to each participant age
df['age_bins'] = pd.cut(df['participant_age.'], bins=bins, labels=labels, right=False)

# Calculating counts
age_counts = df['age_bins'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar')
plt.title('Participant Age Distribution')
plt.xlabel('Age Bins')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Labeling bars with counts
for i, count in enumerate(age_counts):
    plt.text(i, count + 0.5, str(count), ha='center', va='bottom')

plt.show()
```
